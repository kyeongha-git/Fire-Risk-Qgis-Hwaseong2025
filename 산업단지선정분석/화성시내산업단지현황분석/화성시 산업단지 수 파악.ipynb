```python
import pandas as pd
```

```python
df = pd.read_csv("data/전국_공장_등록현황.csv")
df
```

```python
def replace_(x):
    return x.replace(",", "")
```

```python
df['합계'] = df['합계'].apply(replace_)
```

```python
df
```

```python
df['합계'] = df['합계'].astype(int)
```

```python
# 내림차순 정렬
df.sort_values('합계', ascending = False)
```

```python
# 화성시 외 다른 지역 DataFrame
df_sub = df[df['시군구명'] != '화성시']
df_sub
```

```python
# 다른 지역 평균 산업단지 수
avg = df_sub['합계'].mean()
avg
```

```python
# 화성시의 평균 산업단지 수가 다른 지역에 비해 몇 배나 많은지
# 화성시 산업단지 수 / 다른 지역 평균 산업단지 수 = 16.46 -> 화성시는 다른 지역 대비 약 16배 많은 산업단지를 보유함.
12488 / avg
```
